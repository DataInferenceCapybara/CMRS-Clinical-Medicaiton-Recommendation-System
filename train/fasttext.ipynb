```python
cores = multiprocessing.cpu_count()
model = FastText(window=5,
                 vector_size=300,
                 sample=1e-3, 
                 alpha=0.025, 
                 min_count=5,
                 epochs=50,
                 workers=cores-1,
                 seed=42
                )
model.build_vocab(note_token, progress_per=100000)
model.train(note_token, total_examples= model.corpus_count, report_delay=1, epochs=model.epochs)
model.save('../data/Fasttext_Multi_300.mdl')
```
